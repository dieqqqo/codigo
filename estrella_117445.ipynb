```python
#Master BIAS
master_bias= np.mean(np.array([bias1[0].data,bias2[0].data,bias3[0].data,\
bias4[0].data,bias5[0].data,bias6[0].data,bias7[0].data,bias8[0].data,bias9[0].data]), axis=0)

#DARK V
darkE2_Vlist= np.array([darkE2_V1[0].data, darkE2_V2[0].data, darkE2_V3[0].data, darkE2_V4[0].data,\
darkE2_V5[0].data]) - master_bias

#DARK B
darkE2_Blist= np.array([darkE2_B1[0].data, darkE2_B2[0].data, darkE2_B3[0].data, darkE2_B4[0].data,\
darkE2_B5[0].data]) - master_bias

#Master DARK V
masterE2V_dark = np.median(darkE2_Vlist, axis = 0)
darkE2_V_bias= masterE2V_dark + master_bias

#Master DARK B
masterE2B_dark = np.median(darkE2_Blist, axis = 0)
darkE2_B_bias= masterE2B_dark + master_bias

#FLAT V y B

flat_Vlist = np.mean(np.array([flatV1[0].data, flatV2[0].data, flatV3[0].data, flatV4[0].data, flatV5[0].data]), axis=0)

flat_Blist = np.mean(np.array([flatB1[0].data, flatB2[0].data, flatB3[0].data, flatB4[0].data,flatB5[0].data]), axis = 0)

#MASTER FLAT V y B

master_flatV = (flat_Vlist - darkE2_V_bias)/(np.median(flat_Vlist - darkE2_V_bias))

master_flatB = (flat_Blist - darkE2_B_bias)/(np.median(flat_Blist - darkE2_B_bias))
```

```python
## Definimos nuestro kernel de boxcar.

boxcar=(np.vstack([np.ones(3),np.ones(3),np.ones(3)]).T)/6
```

```python
#Aplicamos 3 filtros diferentes a cada imagen profunda (Gaussiano,Laplaciano y Boxcar)
gaussV=scipy.ndimage.gaussian_filter(V,5)
laplaceV=scipy.ndimage.uniform_filter(V)
boxcarV=scipy.ndimage.convolve(V,boxcar)
```

```python
gaussB=scipy.ndimage.gaussian_filter(B,5)
laplaceB=scipy.ndimage.uniform_filter(B)
boxcarB=scipy.ndimage.convolve(B,boxcar)
```
